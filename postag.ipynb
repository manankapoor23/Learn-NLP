# Dr. APJ Abdul Kalam Speech Corpus

A collection of excerpts and quotes from Dr. APJ Abdul Kalam, former President of India and renowned scientist.


```python
paragraph = "Natural Language Processing is a rapidly growing field of artificial intelligence. It enables computers to understand, analyze, and generate human language. Many applications such as I am chatbots, search engines, and voice assistants rely on NLP techniques to process text efficiently"
```

```python
import nltk 
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk.corpus import stopwords
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in set(stopwords.words("english"))]
    x = nltk.pos_tag(words)
    print(x)
```
